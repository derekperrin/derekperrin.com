## Groups
We'll be working a lot with algebraic sets known as _groups_. Formally, a group $G$ is a non-empty set along with a binary operation $\cdot: G \times G \to G$ that satisfies the following:

1. **Identity**: There exists a unique element $e$ in $G$ such that for any $g$ in $G,$ $e \cdot g = g \cdot e = g.$
2. **Associativity**: Let $f, g, h$ be elements of $G.$ Then $f \cdot (g \cdot h) = (f \cdot g) \cdot h).$
3. **Inverses**: For any $g$ in $G$, there exists a unique element $h$ such that $g \cdot h = h \cdot g = e.$ We denote this element $h = g^{-1}.$

Loosely speaking, a group is a set of objects which we can "add" together to get other objects in the group. There exists an element $e$ known as the identity, and it is the "do-nothing" element (i.e. if we "add" it to something else, nothing happens). We can obtain $e$ by "subtracting" any object from itself.
